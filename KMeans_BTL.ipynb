Thuộc tính

Mọi người

* ID: Mã định danh duy nhất của khách hàng

không

* AcceptedCmp3: 1 nếu khách hàng chấp nhận ưu đãi trong chiến dịch thứ 3, 0 nếu không

* AcceptedCmp4: 1 nếu khách hàng chấp nhận ưu đãi trong chiến dịch thứ 4, 0 nếu không

* AcceptedCmp5: 1 nếu khách hàng chấp nhận ưu đãi trong chiến dịch thứ 5, 0 nếu không

Scale dữ liệu

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(data)
```

```python
X_scaled
```

Xây dựng mô hình

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(2, 10), wcss, marker="o")
plt.xlabel("k")
plt.ylabel("WCSS (Inertia)")
plt.title("Elbow Method")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

K_range = range(2, 11)  # thử các số cụm từ 2 đến 10
sil_scores = []

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)  # X_scaled là dữ liệu đã chuẩn hóa
    score = silhouette_score(X_scaled, labels)
    sil_scores.append(score)

# In ra Silhouette score cho từng k
for k, score in zip(K_range, sil_scores):
    print(f"k={k}, Silhouette score={score:.4f}")
# Vẽ biểu đồ
plt.figure(figsize=(8,5))
plt.plot(K_range, sil_scores, marker='o')
plt.title("Silhouette Score theo số cụm k")
plt.xlabel("Số cụm k")
plt.ylabel("Silhouette Score")
plt.xticks(K_range)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()    

```

Cluster 0 – “Moderate Family Buyers” (Nhóm gia đình – chi tiêu vừa phải)
Đặc điểm nhân khẩu học

* Tuổi trung bình: ~55

* Thu nhập: ~39,000 → thuộc nhóm thu nhập trung bình

* Con cái: ~1.2 → thường có gia đình

* Tình trạng hôn nhân: phần lớn đã lập gia đình

* Học vấn: Postgraduate chiếm tỉ lệ cao

Hành vi mua sắm

* Chi tiêu nhìn chung thấp đến trung bình

* Chi tiêu rượu (Wine) thấp (~112)

* Chi tiêu thịt ở mức vừa (~38)

* Ít mua trái cây, đồ ngọt, đồ cao cấp

* Có xu hướng mua deal nhiều hơn (NumDeals ~2.4)

* Tần suất mua online lẫn offline ở mức thấp

Insight

* Đây là nhóm khách hàng trung niên – thu nhập trung bình – tiêu dùng thực dụng, ưu tiên sản phẩm thiết yếu và thích các chương trình khuyến mãi.
